```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
from pylab import *
img = imread('edge1.png')

#Convolve Matrix
con = np.array([[1,-1],[1,-1]])

n=np.shape(img)
f=np.shape(con)
len=n[0]-f[0]+1

#Code for Convolution
arr1 = []
for i in range(0,len):
    for j in range(0,len):
        arr=np.array([])
        for k in range(f[0]):
            for m in range(f[0]):
                if(j>len+1):
                    break;
                arr=np.append(arr,img[i+k][j+m]*con[k][m])
        arr1=np.append(arr1,arr.sum())
arr1=arr1.reshape(len,len)


plt.imshow(img)
plt.title("Original Image")
plt.show()

plt.imshow(arr1)
plt.title("Filtered Image")
plt.show()
```
